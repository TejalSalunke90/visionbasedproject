```python
import numpy as np
import cv2 as cv  # opencv-python-4.5.5.64

#Python 3.9.12

from matplotlib import pyplot as plt
img = cv.imread('input.jpg',cv.COLOR_BGR2GRAY) 

imgplot=plt.imshow(img)
plt.show()

x,y = img.shape[:2]
print(x,y)

imgshape = cv.resize(img, (200, 240))
x2,y2 = imgshape.shape[:2]

print(x2,y2)
img2plot=plt.imshow(imgshape)
plt.show()
```

#Format for cv.FastFeatureDetector_create()

#(	int 	threshold = 10,
#bool 	nonmaxSuppression = true,
#FastFeatureDetector::DetectorType 	type = FastFeatureDetector::TYPE_9_16
#)	

```python


# Initiate FAST object with default values
fast = cv.FastFeatureDetector_create()

# Print all default params
print( "Threshold: {}".format(fast.getThreshold()) )
print( "nonmaxSuppression:{}".format(fast.getNonmaxSuppression()) )
print( "neighborhood: {}".format(fast.getType()) )



# find and draw the keypoints
kp = fast.detect(img,None)
img2 = cv.drawKeypoints(imgshape, kp, None, color=(255,0,0))
cv.imwrite('fast_true.png', img2)
img2plot=plt.imshow(img2)
plt.show()

print( "Total Keypoints with nonmaxSuppression: {}".format(len(kp)) )





```

```python


# Disable nonmaxSuppression
fast.setNonmaxSuppression(0)
kp = fast.detect(imgshape, None)

print( "Total Keypoints without nonmaxSuppression: {}".format(len(kp)) )

img3 = cv.drawKeypoints(imgshape, kp, None, color=(255,0,0))
cv.imwrite('fast_false.png', img3)
img3plot=plt.imshow(img3)
plt.show()
```

```python
# Check the effect of introducing blur to reduce the number of keypoints

ksize = (5, 5)  # Box filter for blurring
imgblr = cv.blur(imgshape, ksize, cv.BORDER_DEFAULT) 

# Initiate FAST object with default values
fast = cv.FastFeatureDetector_create()

# Print all default params
print( "  Threshold: {}".format(fast.getThreshold()) ,"  NonmaxSuppression:{}".format(fast.getNonmaxSuppression()),"  neighborhood: {}".format(fast.getType()))


# find and draw the keypoints
kp = fast.detect(imgblr,None)
img4 = cv.drawKeypoints(imgblr, kp, None, color=(255,0,0))
cv.imwrite('fast_blur_true.png', imgblr)
img4plot=plt.imshow(img4)
plt.show()

print( "Total Keypoints with nonmaxSuppression: {}".format(len(kp)) )
```

#### Effect of Gaussian blur

```python
# Check the effect of introducing Gaussian blur to reduce the number of keypoints

ksize = (5, 5)  # Box filter for blurring
imgGausblr1 = cv.GaussianBlur(imgshape, ksize, cv.BORDER_DEFAULT) 

# Initiate FAST object with default values
fast = cv.FastFeatureDetector_create()

# Print all default params
print( "  Threshold: {}".format(fast.getThreshold()) ,"  NonmaxSuppression:{}".format(fast.getNonmaxSuppression()),"  neighborhood: {}".format(fast.getType()))


# find and draw the keypoints
kp = fast.detect(imgGausblr1,None)
imggblr1 = cv.drawKeypoints(imgGausblr1, kp, None, color=(255,0,0))
cv.imwrite('fast_blur_true.png', imggblr1)
img4plot=plt.imshow(imggblr1)
plt.show()

print( "Total Keypoints with nonmaxSuppression: {}".format(len(kp)) )
```

```python
# load the image, display it to our screen, and construct a list of
# bilateral filtering parameters that we are going to explore
image = cv2.imread(args["image"])
cv2.imshow("Original", image)
params = [(11, 21, 7), (11, 41, 21), (11, 61, 39)]
# loop over the diameter, sigma color, and sigma space
for (diameter, sigmaColor, sigmaSpace) in params:
    # apply bilateral filtering to the image using the current set of
    # parameters
    blurred = cv2.bilateralFilter(image, diameter, sigmaColor, sigmaSpace)
    # show the output image and associated parameters
    title = "Blurred d={}, sc={}, ss={}".format(
        diameter, sigmaColor, sigmaSpace)
    cv2.imshow(title, blurred)
    cv2.waitKey(0)
```
